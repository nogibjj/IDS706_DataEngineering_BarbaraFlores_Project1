```python
# imports
import pandas as pd
import matplotlib.pyplot as plt
from lib import mean_variable, median_variable, count_variable
```

### üìä Results
We can calculate some descriptive statistics such as the mean, median, and total number of records for the following variables:

- "Total_applicants"
- "Employee_count"
- "LinkedIn_Followers""

```python
variable_list = ["Total_applicants", "Employee_count", "LinkedIn_Followers"]
path = '../data/LinkedInTechJobsDataset.csv'

print(" \nSome descriptive statistics: \n")
for i in variable_list:
    print(
        f"The mean of {i} is {round(mean_variable(path, i))}.")
    print(
        f"The median of {i} is {round(median_variable(path, i))}.")
    print(
        f"The count of {i} is {round(count_variable(path, i))}. \n")
```

"However, measures of central tendency often conceal the true distribution of the sample, 
so it is interesting to observe a histogram of the same values."

```python
df = pd.read_csv(path) 
variable = "Total_applicants"
plt.hist(df[[variable]], bins=10,)
plt.xlabel(variable)
plt.ylabel('frequency')
plt.title('Histogram of {0} per job posting'.format(variable))
plt.show()
plt.clf()
```
